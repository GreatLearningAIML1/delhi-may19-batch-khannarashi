```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_1samp, wilcoxon, ttest_rel, ttest_ind
from statsmodels.stats.power import ttest_power
```

```python
Titan_IC = pd.read_excel("Titan_Insurance_.xlsx", index_col=0)
Titan_IC
```

```python
Titan_IC['Old Scheme (in thousands)'] = Titan_IC['Old Scheme (in thousands)'] * 1000
Titan_IC['New Scheme (in thousands)'] = Titan_IC['New Scheme (in thousands)'] * 1000
```

1. Find the mean of old scheme and new scheme column.

```python
mean_old = Titan_IC['Old Scheme (in thousands)'].mean()
mean_new = Titan_IC['New Scheme (in thousands)'].mean()
print(mean_old, mean_new)
```

2. Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs?

```python
# Null Hypothesis: The scheme has not made a significant difference (μ1 = μ2). 
# Alternate Hypothesis: The scheme has made a significant difference (μ1 ≠ μ2).

#Paired sample t-test
t_statistic, p_value = ttest_rel(Titan_IC['Old Scheme (in thousands)'], Titan_IC['New Scheme (in thousands)'])
print(t_statistic, p_value)
```

```python
# Check if p_value is greater than significance value (market value: 0.05) or not
if (p_value < 0.05):
    print("Null Hypothesis is rejected.")
else:
    print("Failed to reject Null Hypothesis.")
```

3. What conclusion does the test (p-value) lead to?

```python
# Conclusion: The scheme has not made a significant difference, i.e. μ1 = μ2!
```

b) What is the p- value of the hypothesis test if we test for a difference of $5000?

```python
# Null Hypothesis: The scheme has not made a break even, i.e., (μ1 - μ2) < 5000. 
# Alternate Hypothesis: The scheme has not made a break even, i.e., (μ1 - μ2) >= 5000.

t_statistic, p_value = ttest_rel(Titan_IC['Old Scheme (in thousands)'], Titan_IC['New Scheme (in thousands)'])
print(t_statistic, p_value)
```

```python
# Check if p_value is greater than significance value (market value: 0.05) or not
if (p_value < 0.05):
    print("Null Hypothesis is rejected.")
else:
    print("Failed to reject Null Hypothesis.")
```
