```python
# Importing the necessary packages

import pandas as pd
import numpy as np
import tensorflow as tf
import keras as keras
```

# 1. Read the dataset

```python
from google.colab import drive
drive.mount('/gdrive')
```

```python
# Reading the dataset

bank_cust_data = pd.read_csv('/gdrive/My Drive/Colab Notebooks/Churn.csv')
bank_cust_data.head()
```

# 2. Drop the columns which are unique for all users like IDs

```python
# Dropping irrelevant columns

bank_cust_data.drop(["RowNumber","CustomerId","Surname"], axis=1, inplace=True)
bank_cust_data.columns
```

# 3. Distinguish the feature and target set

```python
# Before distinguishing, converting the string columns to numerical columns using Label Encoder and creating dummies

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
bank_cust_data['Geography']= label_encoder.fit_transform(bank_cust_data['Geography'])
bank_cust_data['Gender']= label_encoder.fit_transform(bank_cust_data['Gender'])

bank_cust_data = pd.get_dummies(bank_cust_data, columns = ['Geography'])
bank_cust_data.head()
```

```python
bank_cust_data.columns
```

```python
Y = bank_cust_data['Exited']
X = bank_cust_data.drop(['Exited'], axis = 1)
```

# 4. Divide the data set into train and test sets

```python
# Splitting the dataframe into training and test set in the ratio of 70:30 

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 99)
```

# 5. Normalize the train and test data

```python
# Normalizing the data

from sklearn.preprocessing import Normalizer
Normalize = Normalizer()
X_train_norm =  Normalize.fit_transform(X_train) 
X_test_norm =  Normalize.fit_transform(X_test)
```

```python
Y_train = keras.utils.to_categorical(Y_train)
Y_test = keras.utils.to_categorical(Y_test)
```

```python
print(X_train_norm.shape, X_test_norm.shape, Y_train.shape, Y_test.shape)
```

# 6. Initialize & build the model

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout, Flatten, Reshape, BatchNormalization
from keras.layers import Convolution2D, MaxPooling2D
```

```python
#Clear out tensorflow memory
keras.backend.clear_session()

# Define Model
bank_model = Sequential()
bank_model.add(BatchNormalization())

#Adding 1st dense layer
bank_model.add(Dense(100, activation='sigmoid', input_shape=(12,)))

#Add 2nd hidden layer
bank_model.add(Dense(60, activation='sigmoid'))

#Adding OUTPUT layer
bank_model.add(Dense(2, activation='softmax'))
```

```python
#Compile the model
bank_model.compile(optimizer = keras.optimizers.SGD(lr=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
bank_model.fit(X_train_norm, Y_train, epochs=20, validation_data = (X_test_norm, Y_test), batch_size=50)
```

# 7. Optimize the model

```python
# Define Model
bank_model_2 = Sequential()
bank_model_2.add(BatchNormalization())

#Add 1st hidden layer
bank_model_2.add(Dense(30, input_shape=(12,), activation='relu'))
bank_model_2.add(BatchNormalization())

#Add 2nd hidden layer
bank_model_2.add(Dense(20, kernel_initializer='normal', activation='relu'))
bank_model_2.add(BatchNormalization())

#Add 3rd hidden layer
bank_model_2.add(Dense(8, kernel_initializer='normal', activation='relu'))
bank_model_2.add(BatchNormalization())

#Add 4th hidden layer
bank_model_2.add(Dense(5, kernel_initializer='normal', activation='relu'))
bank_model_2.add(BatchNormalization())

#Add 5th hidden layer
bank_model_2.add(Dense(3, kernel_initializer='normal', activation='relu'))
bank_model_2.add(BatchNormalization())

bank_model_2.add(Dropout(0.5))

#Adding OUTPUT layer
bank_model_2.add(Dense(2, activation='sigmoid'))

#Compile the model
bank_model_2.compile(optimizer = keras.optimizers.SGD(lr=0.003), loss='binary_crossentropy', metrics=['accuracy'])

bank_model_2.fit(X_train_norm, Y_train, epochs=50, validation_data = (X_test_norm, Y_test), batch_size=50)
```

# 8. Predict the results using 0.5 as a threshold

```python
Y_pred = bank_model_2.predict(X_test_norm)
```

```python
Y_pred = (Y_pred>0.5)
Y_pred
```

# 9. Print the Accuracy score and confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
ConfusionMatrix = confusion_matrix(Y_test.argmax(axis=1), Y_pred.argmax(axis=1))
AccuracyScore = accuracy_score(Y_test, Y_pred)
```
